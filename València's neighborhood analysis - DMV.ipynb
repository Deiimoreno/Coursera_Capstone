# 1. Business Problem - Which restaurant will fit better in València?

### Background

València, a sunny city placed in the penninsula ibérica's east coast, is one of the best cities to live in Europe, as some reports from important brands have said, i.e. this article from [globexs.com](https://www.globexs.com/blog/living-valencia-why-it-better-live-valencia-madrid-or-barcelona). In order to get in place, bunches of restaurants have already settled in Valencia, also local franchise restaurants are raising and becoming very popular in the city and all the region. Although is not the Spain's most populated city (instead it's the third), the city receives almost 2 million tourists every year, read this article from [Levante](https://www.levante-emv.com/valencia/2019/02/07/valencia-supera-millones-turistas-precios/1831764.html) (in spanish)  - and this numbers have just started to grow up! 

Therefore, there are plenty of reasons to open a business in Valencia. On the one hand you have a 1.5 million habitants city so locals will give you an opportunity to grow up. On the other hand, you have the high amount of tourists coming to the city and bringing you the chance to serve them the best quality service that you can.

### Problem

Supose some investor wants to open a restaurant in València. Capital is available but there's no preference for choosing a topic (i.e. mexican, japanese, tapas, greek). Instead of giving importance to which topic would be chosen, stakeholders were more interested in reporting an analytic of which category will fit the most in each neighborhood - but where? How? Which restaurant will suit better? 

These are the key questions about the problem.
